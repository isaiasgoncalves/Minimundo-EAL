```python
import pandas as pd
```

```python
data = pd.read_csv("acidentes2025_todas_causas_tipos.csv", encoding='latin-1', sep=";")
data.shape
```

```python
# Create a list of the columns to check
data_cleaned = data.dropna()
print(data_cleaned.shape)
```

```python
x = data_cleaned.shape[0]
y =data_cleaned[data_cleaned["ano_fabricacao_veiculo"]<=0].shape[0]
data_cleaned[data_cleaned["ano_fabricacao_veiculo"]<=0].shape[0]
```

```python
data_cleaned = data_cleaned[data_cleaned["ano_fabricacao_veiculo"]>0]
```

```python
important_col = [ "causa_acidente", "tipo_acidente", "classificacao_acidente", "estado_fisico", "condicao_metereologica"]
```

```python
dicionario_score =[]
for col in important_col:
    dicionario_score_cat = (data_cleaned.groupby(col)['id'].count()/data_cleaned.shape[0]).to_dict()
    dicionario_score_cat =sorted(dicionario_score_cat.items(), key=lambda item:item[1], reverse=True)
    
    dicionario_score_cat = [(f"{col}.{elem[0]}", elem[1]) for elem in dicionario_score_cat]
    dicionario_score+=dicionario_score_cat
    
```
